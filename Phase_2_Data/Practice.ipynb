```python
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data
    
```

```python
def clean_data(data):
    text_to_num = {"one": 1, "two" : 2, "three" : 3, "four" : 4,"five" :5 }
    cln_data = []
    unq_user = set()
    for user in data:
        # handling inconsistency -- rating
        raw_rating = str(user["rating"]).strip().lower()
        if (raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating
        # print(raw_rating)

        # handling the missing values
        if (user.get("age") == None):
            user["age"] = None
        # handling for duplications
        if( user["name"] in unq_user):
            continue
        unq_user.add(user["name"])
        cln_data.append(user)
        
    return cln_data
```

```python
b= clean_data(data)
for i in b:
    print(i)
```

```python
sum = 0
count = 0
poor = 0
for i in b:
    int_rating  = float(i["rating"])
    count +=1
    sum += int_rating
    if (int_rating < 3):
        poor += 1

```

```python
print("total reviews are", count)
print("the average rating is: ",sum/count)
print("the percentage of the poor rating is: ",(poor/count)*100)
    
```

```python
def get_recommendation(b):
    recommendation = []
    for i in b:
        curr_recom = {}
        curr_recom["name"] = i["name"]
        if(float(i["rating"] ) >= 4):
            curr_recom["brand"] = "Apple"
        else:
            curr_recom["brand"] = "Samsung"
        
        recommendation.append(curr_recom)
    return recommendation
```
