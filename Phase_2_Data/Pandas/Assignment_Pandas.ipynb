```python
import pandas as pd
df = pd.read_csv("IRIS.csv")
```

```python
print(type(df))
df.head(10)
df.shape
df.describe()
```

```python
df.query(' petal_length > 4.5 and species =="Iris-virginica" ')
# df[ (df["petal_length"] > 4.5) & (df["species"] =="Iris-virginica")]
```

```python
avg_sepal_length = df.groupby(["species"])["sepal_length"].mean()
max_petal_width = df.groupby(["species"])["petal_width"].max()
std_petal_width = df.groupby(["species"])["petal_width"].std()
print(f"std{std_petal_width}\n and max{max_petal_width}\n abd {avg_sepal_length}")

df["petal_ratio"] = df["petal_length"]/df["petal_width"]
df.groupby("species")["petal_ratio"].mean()
```
