```python
from bs4 import BeautifulSoup
```

```python
with open("scraped_data/data_1.html","r") as f:
    html_content = f.read()
```

```python
soup = BeautifulSoup(html_content,"lxml")
```

```python
all_h3 = soup.find_all("h3")
# for h3 in all_h3:
#     h3.get_text()
```

```python
b_list = []

for h3 in all_h3:
    # Get the current book's article
    article = h3.find_parent("article", class_="product_pod")

    # Full book title
    book_name = h3.find("a")["title"]

    # Book price
    price = article.find("p", class_="price_color").get_text(strip=True)

    # Availability
    availability = article.find("p", class_="instock availability").get_text(strip=True)

    # Rating
    rating = article.find("p", class_="star-rating")["class"][1]
    b_dict = {
        "book name": book_name,
        "book price": price,
        "availability": availability,
        "rating": rating
    }
    b_list.append(b_dict)
    print(f"Book Name   : {book_name}")
    print(f"Price       : {price}")
    print(f"Availability: {availability}")
    print(f"Rating      : {rating}")
    print("-" * 50)
```

```python
for i in b_list:
    print(i)
```
