```python
import json
```

```python

def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
        
    return data
```

```python
load_data("store_data.json")
```

```python
data = load_data("store_data.json")
print(type(data))
for i in data:
    print(i)
```

```python
 # clean and structure the data
def clean_data(data):
    # structure the rating  - data consistency
    text_to_num = {"one": 1, "two":2, "three": 3, "four": 4, "five": 5}
    cleaned_data = []
    unique_users = set()
    for user in data:
        raw_rating = str(user["rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        # print(raw_rating) # printing the raw data
        user["rating"] = raw_rating # updating the data

        # handling the missing values
        raw_age = user.get("age")
        if (raw_age == None):
            user["age"] = None

        # Deduplication 
        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data
        
```

```python
a=clean_data(data)
for i in a:
    print(i)
```
