```python
# fast calculation 
size = 100_000
py_list = list(range(size))
start_time = time.time()
sq_list = [i **2 for i in py_list]
end_time = time.time()
py_time = end_time - start_time
print(f"the time taken by py list is: {end_time - start_time}")
#  np calculation
np_arr = np.array(py_list)
start_time = time.time()
# vectorizition
sq_arr = np_arr **2
end_time = time.time()
np_time = end_time - start_time
print(f"the time taken by np list is: {end_time - start_time}")
time_difference = (py_time / np_time ) *100
print(time_difference)
```

```python
# memory less taken + np is stored on contiguous memory location
import sys
py_size = sys.getsizeof(py_list) * len(py_list)
print(f"the size of the py list is: {py_size}")
np_size = np_arr.nbytes
print(f"the size of the np array is: {np_size}")
memory_diff = (py_size / np_size )*100
print(memory_diff)
```

```python
# Operations on Numpy
            # 1.Reshaping
b3 = np.array(list(range(1,21)))
print(b3, b3.dtype,b3.shape,b3.ndim)
# from flatten to matrix 2d
b4 = b3.reshape(5,4)
print(b4, b4.dtype,b4.shape,b4.ndim)

b5 = b4.flatten()
print(b5,b5.shape)

            # 2. Indexing

# Normal index
print(b4[1][2]) # 6
# for i in b4:
#     print(i)
# Fancy Indexing --- very easy b/c of vectorization  -- see it as whole rather than taversing 
f_index = [2,0,5]
print(b5[f_index])

# Boolean Indexing  --- basesd on condition true -- print or access the index
print(b5 [ b5 % 2 == 0 ])

            # 3. Slicing
print(b5[2:8]) # 3  4  5  6  7  8
print(b5[2:])
print(b5[:8])
print(b5[2:20:2])
print(b5[::-1])

#  copy VS view slicing in numpy is similar to list but big differenc in terms of interal works
# py list copy the list but in numpy array view is created which showing the original menas no agin creating like list
#  copy VS view

c_list =  [1,3,5,6,1,2]
print(c_list)
c_sub = c_list[0:4]
print(c_sub)
c_sub[1] = 300
print(c_sub)

print(f"agian the original : {c_list}")

c_np = np.array(c_list)
c_np_sub = c_np[0:4]   # .copy() this will not copy the entir but the copy of the slice view into the variable
print(c_np_sub)
c_np_sub [0] = 100
print("the updated sub: ",c_np_sub)
print("the updated origial np: ",c_np)
```

```python
# Multi-dimensional arrays
arr2D = np.array([[3,6,2,6] ,[2,7,8,3],[2,6,2,7]])
print(arr2D,arr2D.dtype)
print(np.sum(arr2D))

sum_columns = np.sum(arr2D, axis = 0)
print(sum_columns)

sum_rows = np.sum(arr2D, axis= 1)
print(sum_rows)

# index slicing through axis
# accessing 7 8 
print(arr2D[1:3 , 1])

#      3D Array
arr3D = np.array([ [[1,2] , [3,4]] , [[5,6],[7,8]] ])
print(arr3D ,arr3D.shape)
print(f"accessing the 5 -- 1,0,0 the item is:  {arr3D[1,0,0]}")
#  index slicing through 3D
print(arr3D[ : , 0 , : ] )
 
```

```python
# vector Normalization
arr = np.array([[2,4,9],
                [7,5,1],
               [1,2,4]])
mean = np.mean(arr)
print(mean)
st_dev = np.std(arr)
print(st_dev)
normalize = (arr - mean)/st_dev
print(normalize)
print(np.sum(normalize))
```
