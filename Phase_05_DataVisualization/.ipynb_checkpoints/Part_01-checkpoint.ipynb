```python
# Real data

# Oscar Winner Movies
oscar_movies = [
"The Dark Knight",
"The Hurt Locker",
"The King's Speech",
"The Artist",
"Argo"]
# Oscar Moives 
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M
non-oscar_revenue = [378,2788,829,185,275]
# Plot
plt.plot(years,oscar_revenue)
plt.plot(years,non-oscar_revenue)

plt.title("oscar Moives Revenue in each Year")
plt.xlabel("years")
plt.ylabel("Revenue in Millions Dollers")
plt.show()

```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

sales = [120, 150, 170, 160, 190, 220,
         240, 230, 210, 250, 280, 300]

plt.plot(months,sales)
plt.title("Sales in Months")
plt.xlabel("months")
plt.ylabel("sales")
plt.show()
```
